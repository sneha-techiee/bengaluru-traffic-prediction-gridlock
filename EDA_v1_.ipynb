```python
#Geohash-lat/lon
import pygeohash as pgh

def decode(gh):
    try:
        lat, lon = pgh.decode(gh)
        return pd.Series([lat, lon])
    except:
        return pd.Series([0, 0])

train[['lat', 'lon']] = train['geohash'].apply(decode)
test[['lat', 'lon']] = test['geohash'].apply(decode)
```

```python
import pandas as pd
import numpy as np
import pygeohash as pgh
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
import lightgbm as lgb
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv("train.csv")
test  = pd.read_csv("test.csv")

# Missing value fills
train['Temperature'] = train['Temperature'].fillna(train['Temperature'].median())
test['Temperature']  = test['Temperature'].fillna(train['Temperature'].median())

weather_mode = train['Weather'].mode()[0]
train['Weather'] = train['Weather'].fillna(weather_mode)
test['Weather']  = test['Weather'].fillna(weather_mode)

train['RoadType'] = train['RoadType'].fillna('Unknown')
test['RoadType']  = test['RoadType'].fillna('Unknown')

for col in ['LargeVehicles', 'Landmarks']:
    m = train[col].mode()[0]
    train[col] = train[col].fillna(m)
    test[col]  = test[col].fillna(m)
```

```python
for df in [train, test]:
    df['hour']       = df['timestamp'].apply(lambda x: int(x.split(':')[0]))
    df['minute']     = df['timestamp'].apply(lambda x: int(x.split(':')[1]))
    df['time_block'] = df['hour'] * 4 + df['minute'] // 15
    df['hour_sin']   = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos']   = np.cos(2 * np.pi * df['hour'] / 24)
    df['is_morning_peak'] = df['hour'].isin([7,8,9]).astype(int)
    df['is_evening_peak'] = df['hour'].isin([17,18,19]).astype(int)
    df['is_night']        = df['hour'].isin([0,1,2,3,4,5]).astype(int)
```

```python
def decode_gh(gh):
    lat, lon = pgh.decode(gh)
    return float(lat), float(lon)

for df in [train, test]:
    coords = df['geohash'].apply(decode_gh)
    df['lat'] = coords.apply(lambda x: x[0])
    df['lon'] = coords.apply(lambda x: x[1])
    df['geo4'] = df['geohash'].str[:4]
    df['geo3'] = df['geohash'].str[:3]
```

```python
# Compute ALL stats on train only
global_mean = train['demand'].mean()

def smoothed_mean(df_train, df_target, group_cols, alpha=10):
    stats = df_train.groupby(group_cols)['demand'].agg(['mean','count'])
    stats['sm'] = (stats['mean']*stats['count'] + global_mean*alpha) / (stats['count']+alpha)
    key = '_'.join(group_cols) + '_mean'
    return stats['sm'].rename(key), key

# geo_mean (geohash level)
geo_sm, _ = smoothed_mean(train, train, ['geohash'])
for df in [train, test]:
    df['geo_mean'] = df['geohash'].map(geo_sm).fillna(global_mean)

# geo4_mean (neighbourhood level)
geo4_sm, _ = smoothed_mean(train, train, ['geo4'])
for df in [train, test]:
    df['geo4_mean'] = df['geo4'].map(geo4_sm).fillna(global_mean)

# geo3_mean (zone level)
geo3_sm, _ = smoothed_mean(train, train, ['geo3'])
for df in [train, test]:
    df['geo3_mean'] = df['geo3'].map(geo3_sm).fillna(global_mean)

# geo_hour_mean — SMOOTHED (this is what was killing you before)
geo_hour_stats = train.groupby(['geohash','hour'])['demand'].agg(['mean','count'])
geo_hour_stats['sm'] = (
    (geo_hour_stats['mean']*geo_hour_stats['count'] + global_mean*10)
    / (geo_hour_stats['count']+10)
)
for df in [train, test]:
    df['geo_hour_mean'] = df.set_index(['geohash','hour']).index.map(
        geo_hour_stats['sm']
    ).fillna(global_mean).values
```

```python
# Build lookup from day 48 only
lag_lookup = (
    train[train['day']==48]
    .groupby(['geohash','timestamp'])['demand']
    .mean()
    .reset_index()
    .rename(columns={'demand':'lag_demand'})
)

# Day 48 train rows: use geo_mean as lag (no day 47 data)
train['lag_demand'] = train['geo_mean'].copy()
train['lag_missing'] = 0

# Day 49 val rows: merge actual day48 demand
day49_idx = train['day'] == 49
day49_with_lag = train[day49_idx].drop(columns=['lag_demand','lag_missing']).merge(
    lag_lookup, on=['geohash','timestamp'], how='left'
)
day49_with_lag['lag_missing'] = day49_with_lag['lag_demand'].isna().astype(int)
day49_with_lag['lag_demand'] = day49_with_lag['lag_demand'].fillna(
    day49_with_lag['geo_mean']
)
train.loc[day49_idx, 'lag_demand'] = day49_with_lag['lag_demand'].values
train.loc[day49_idx, 'lag_missing'] = day49_with_lag['lag_missing'].values

# Test rows
test = test.merge(lag_lookup, on=['geohash','timestamp'], how='left')
test['lag_missing'] = test['lag_demand'].isna().astype(int)
test['lag_demand'] = test['lag_demand'].fillna(test['geo_mean'])

# Ratio feature
for df in [train, test]:
    df['lag_ratio'] = df['lag_demand'] / (df['geo_mean'] + 1e-5)
```

```python
cat_cols = ['RoadType', 'Weather', 'LargeVehicles', 'Landmarks']
for col in cat_cols:
    le = LabelEncoder()
    combined = pd.concat([train[col], test[col]]).astype(str)
    le.fit(combined)
    train[col] = le.transform(train[col].astype(str))
    test[col]  = le.transform(test[col].astype(str))
```

```python
FEATURES = [
    'hour', 'minute', 'time_block', 'hour_sin', 'hour_cos',
    'is_morning_peak', 'is_evening_peak', 'is_night',
    'lat', 'lon',
    'NumberofLanes', 'Temperature',
    'RoadType', 'Weather', 'LargeVehicles', 'Landmarks',
    'geo_mean', 'geo4_mean', 'geo3_mean', 'geo_hour_mean',
    'lag_demand', 'lag_missing', 'lag_ratio'
]

X_tr  = train[train['day']==48][FEATURES]
y_tr  = train[train['day']==48]['demand']
X_val = train[train['day']==49][FEATURES]
y_val = train[train['day']==49]['demand']

model = lgb.LGBMRegressor(
    objective='regression_l1',
    n_estimators=3000,
    learning_rate=0.03,
    num_leaves=255,
    min_child_samples=20,
    feature_fraction=0.8,
    bagging_fraction=0.8,
    bagging_freq=1,
    reg_alpha=0.1,
    reg_lambda=0.1,
    random_state=42,
    n_jobs=-1
)

model.fit(
    X_tr, y_tr,
    eval_set=[(X_val, y_val)],
    callbacks=[lgb.early_stopping(100), lgb.log_evaluation(200)]
)

preds = model.predict(X_val)
print("DAY-BASED R²:", r2_score(y_val, preds))
```
